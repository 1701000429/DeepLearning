```python
#机器学习--线性回归
#authored by waitingfree 2021、12、7
#参考https://www.bilibili.com/video/BV1Xt411s7KY?p=5&spm_id_from=pageDriver
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt #导入画函数曲线的库
```

```python
#当有多个特征时，特征之间的数据差异可能很大，作归一化的好处可以使得梯度下降更快地收敛
def normalize_feature(data):
    #  z=(x-均值)/方差 ,服从标准正态分布，数据分布在[-1,1]
    #  z=(x-min)/(max-min),数据分布在[0,1]
    return (data-data.mean())/data.std()
```

```python
#假设函数（待拟合函数） H(x)=W0+W1*x1+W2*x2
def y_(theta,x,x2):
    return theta[0,0]+theta[1,0]*x+theta[2,0]*x2
```

```python
#损失函数定义:均方误差
def costFunction(X,y,theta):
    #不用for循环累加了，使用矩阵运算得到97*1的向量，然后用numpy求和
    inner=np.power(X@theta-y,2)
    return np.sum(inner)/(2*len(X))
```

```python
#定义梯度下降函数，迭代求解最优的theta1，theta2使得代价函数最小
def gradientDescent(X,y,theta,alpha,iters):
    costList=[]
    for i in range(iters):
        #向量化表示，通过变换得到矩阵形式的迭代公式
        theta=theta-(X.T@(X@theta-y))*alpha/len(X)
        cost=costFunction(X,y,theta)
        costList.append(cost)
    print("最终的损失为：")
    print(cost)
    #返回我们的训练参数
    return theta,costList
```

```python
#导入数据  ex1data1.txt 两列数据（人口数据，利润）
#使用pandas读数据，api为pandas.read_csv(文件路径,分隔符，列名（第几行）,...,names=[标签])
data=pd.read_csv('ex1data2.txt',names=['size','num','price'])
data=normalize_feature(data)
data.plot.scatter("size","price",label="size")
plt.show()#图表可视化
```

```python
#定义数据结构，根据txt文件，构造矩阵X W y
#插入一列数据x0(全为1)，使用的api为insert(插入的列索引,插入列的标签，插入列的值)
data.insert(0,'ones',1)
#切片使用的api为iloc[行，列]
X=data.iloc[:,0:-1]
X=np.array(X)
y=data.iloc[:,3:4]
y=np.array(y)
y.reshape(47,1)
#定义待训练参数，一共两个参数
theta=np.zeros((3,1))
```

```python
#观察学习率对收敛速度的影响
#定义学习率，迭代次数。进行训练，得到两个参数theta1，theta2
alpha=0.02
iters=200
index = np.arange(iters)
fig2,ax2=plt.subplots()
alphaList=[0.00003,0.003,0.03,0.001,0.1,1]
for iterator in alphaList:
    theta,costList=gradientDescent(X,y,theta,iterator,iters)
    ax2.plot(index,costList,label=iterator)
    ax2.legend()
ax2.set(xlabel='iters',ylabel='costs',title='gradientDescent')
plt.show()
```

```python
#验证theta参数的拟合效果
x=np.linspace(X[:,-2].min(),X[:,-2].max(),200)
x2=np.linspace(X[:,-1].min(),X[:,-1].max(),200)
y_hat=y_(theta,x,x2)
print(theta)
fig,ax=plt.subplots()
ax.scatter(X[:,1:2],y,label="traing data:area")#离散点
ax.plot(x,y_hat,'r',label="predict")#连续曲线
ax.legend()#显示label
ax.set(xlabel='area',ylabel='money')
```
