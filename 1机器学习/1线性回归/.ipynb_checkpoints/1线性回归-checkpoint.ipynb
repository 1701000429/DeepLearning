```python
#机器学习--线性回归
#authored by waitingfree 2021、12、7
#参考https://www.bilibili.com/video/BV1Xt411s7KY?p=5&spm_id_from=pageDriver
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt #导入画函数曲线的库
```

```python
#导入数据  ex1data1.txt 两列数据（人口数据，利润）
#使用pandas读数据，api为pandas.read_csv(文件路径,分隔符，列名（第几行），)
data=pd.read_csv('ex1data1.txt',names=['population','profit'])
data.head()#查看前五行数据
data.tail()#查看后五行数据
data.describe()#查看数据的描述
data.plot.scatter("population","profit")
plt.show()#图表可视化
```

```python
#定义数据结构，根据txt文件，构造矩阵X W y
#插入一列数据x0（全为1），使用的api为insert(插入的列索引,插入列的标签，插入列的值)
data.insert(0,'ones',1)
#切片使用的api为iloc[行，列]
X=data.iloc[:,0:2]
X=np.array(X)
y=data.iloc[:,2:3]
y=np.array(y)
y.shape
#定义待训练参数，一共两个参数
theta=np.zeros((2,1))
```

```python
#假设函数（待拟合函数） H(x)=W0+W1*x
def y_(theta,x):
    return theta[0,0]+theta[1,0]*x
```

```python
#损失函数定义:均方误差
def costFunction(X,y,theta):
    #不用for循环累加了，使用矩阵运算得到97*1的向量，然后用numpy求和
    inner=np.power(X@theta-y,2)
    return np.sum(inner)/(2*len(X))
```

```python
#定义梯度下降函数，迭代求解最优的theta1，theta2使得代价函数最小
def gradientDescent(X,y,theta,alpha,iters):
    costList=[]
    for i in range(iters):
        #向量化表示，通过变换得到矩阵形式的迭代公式
        theta=theta-(X.T@(X@theta-y))*alpha/len(X)
        cost=costFunction(X,y,theta)
        costList.append(cost)
    print("最终的损失为：")
    print(cost)
    #返回我们的训练参数
    return theta,costList
```

```python
#定义学习率，迭代次数。进行训练，得到两个参数theta1，theta2
alpha=0.02
iters=2000
theta,costList=gradientDescent(X,y,theta,alpha,iters)
#画图
index = np.arange(iters)
fig,ax=plt.subplots()
ax.plot(index,costList)
ax.set(xlabel='iters',ylabel='costs',title='gradientDescent')
plt.show()
```

```python
#验证theta参数的拟合效果
x=np.linspace(X[:,-1].min(),X[:,-1].max(),200)
y_hat=y_(theta,x)
fig,ax=plt.subplots()
ax.scatter(X[:,1],y,label="traing data")#离散点
ax.plot(x,y_hat,'r',label="predict")#连续曲线
ax.legend()#显示label
ax.set(xlabel='population',ylabel='money')
```
