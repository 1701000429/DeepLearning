```python
#机器学习--逻辑回归
#authored by waitingfree 2021、12、7
#参考https://www.bilibili.com/video/BV1Xt411s7KY?p=5&spm_id_from=pageDriver
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt #导入画函数曲线的库
```

```python
#读数据，数据格式（成绩1，成绩2，是否上大学）
path="ex2data1.txt"
data=pd.read_csv(path,names=['Exam1','Exam2','Accepted'])
data.tail()
#matplotlib做可视化
fig,ax = plt.subplots()
ax.scatter(data[data['Accepted']==0]['Exam1'],data[data['Accepted']==0]['Exam2'],c='r',marker='x',label='y=0')
ax.scatter(data[data['Accepted']==1]['Exam1'],data[data['Accepted']==1]['Exam2'],c='b',marker='o',label='y=1')
ax.legend()
ax.set(xlabel='exam1',
          ylabel='exam2')
plt.show()
```

```python
#定义激活函数 logistic
def sigmoid(z):
    return 1/(1+np.exp(-z));
```

```python
#假设函数（待拟合函数） H(x)=W0+W1*x
def y_(theta,x,x2):
    return sigmoid(theta[0,0]+theta[1,0]*x+theta[2,0]*x2)
```

```python
#损失函数定义:对数几率 J=-1/m [y*log(y_)   +   (1-y)*log(1-y_) ]
#其中  y_=sigmoid（ theta0+theta1*x1+theta2*x2）   向量化则是 S（ X@theta）
def costFunction(X,y,theta):
    y_=sigmoid(X@theta)
    #定义损失函数的两部分
    true_al=y*np.log(y_)
    false_al=(1-y)*np.log(1-y_)
    #返回值
    return np.sum(true_al+false_al)/len(y)*(-1)
```

```python
#定义梯度下降函数，迭代求解最优的theta1，theta2使得代价函数最小
def gradientDescent(X,y,theta,alpha,iters):
    costList=[]
    for i in range(iters):
        A=sigmoid(X@theta)
        #向量化表示，通过变换得到矩阵形式的迭代公式
        theta=theta-(X.T@(A-y))*alpha/len(X)
        cost=costFunction(X,y,theta)
        costList.append(cost)
        if(i%1000==0):
            print(cost)
    print("最终的损失为：")
    print(cost)
    #返回我们的训练参数
    return costList,theta
```

```python
def predict(X,theta):
    probility=sigmoid(X@theta)
    return [1 if x>=0.5 else 0 for x in probility]
```

```python
#定义数据结构
#定义数据结构，根据txt文件，构造矩阵X W y
#插入一列数据x0（全为1），使用的api为insert(插入的列索引,插入列的标签，插入列的值)
data.insert(0,'ones',1)
#切片使用的api为iloc[行，列]
X=data.iloc[:,0:-1]
X=np.array(X)
y=data.iloc[:,-1]
y=np.array(y)
#定义待训练参数，一共两个参数
theta=np.zeros((3,1))
y=y.reshape(len(y),1)
y.shape
```

```python
costFunction(X,y,theta)
#定义训练参数
alpha=0.004
iters=2000000
```

```python
#训练
costList,theta_final=gradientDescent(X,y,theta,alpha,iters)
print(theta_final)
```

```python
#预测(根据训练样本)
y_hat=np.array(predict(X,theta_final))
y_pre=y_hat.reshape(len(y_hat),1)
acc=np.mean(y_pre==y)
print(acc)
```

```python
#绘制直线
coef1 = - theta_final[0,0] / theta_final[2,0]
coef2 = - theta_final[1,0] / theta_final[2,0]
x=np.linspace(X[:,1:2].min(),X[:,1:2].max(),300)
f=coef1+coef2*x
#matplotlib做可视化
fig,ax = plt.subplots()
ax.scatter(data[data['Accepted']==0]['Exam1'],data[data['Accepted']==0]['Exam2'],c='r',marker='x',label='y=0')
ax.scatter(data[data['Accepted']==1]['Exam1'],data[data['Accepted']==1]['Exam2'],c='b',marker='o',label='y=1')
ax.legend()
ax.set(xlabel='exam1',
          ylabel='exam2')
ax.plot(x,f,c='g')
plt.show()
```
